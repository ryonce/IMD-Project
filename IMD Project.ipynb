## Filtering Data - Missing Values

```python
# Replace \N values

basics = basics.replace({'\\N':np.nan})
ratings = ratings.replace({'\\N':np.nan})
akas = akas.replace({'\\N':np.nan})
```

```python
# Exclude any movie with missing values for genre or runtime

print(f'There are {basics.isna().sum().sum()} missing values.')

print(basics.isna().sum())
```

```python
# Drop genres

basics.dropna(subset = 'genres', inplace = True)
```

```python
# Drop runtimeMinutes

basics.dropna(subset = 'runtimeMinutes', inplace = True)
```

```python
# Check values again

print(f'There are {basics.isna().sum().sum()} missing values.')

print(basics.isna().sum())
```

## Filtering Data - Movie

```python
# Include only full-length movies (titleType = "movie").

movie = basics['titleType'] == 'movie'
basics = basics[movie]
basics.head()
```

## Filtering Data - Documentary

```python
# Include only fictional movies (not from documentary genre)

is_documentary = basics['genres'].str.contains('documentary', case=False)
basics = basics[~is_documentary]
basics.head()
```

## Filtering Data - StartYear 2000-2022

```python
basics['startYear'].astype('float')
years = basics['startYear'] >= 2000
basics = basics[years]
basics.head()
```

## Filtering Data - US Only

```python
# Create filter in akas DF

usa = akas['region'] == 'US'
akas = akas[usa]
akas.head()
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe

keepers = basics['tconst'].isin(akas['titleId'])
keepers
```

```python
# Created final DF

basics = basics[keepers]
basics.head()
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe

keepers2 = ratings['tconst'].isin(akas['titleId'])
keepers2
```

```python
# Created final DF

ratings = ratings[keepers2]
ratings.head()
```
